```python
import pandas as pd
```

```python
dados = pd.read_csv('dados/aluguel_residencial.csv', sep = ';')
```

```python
dados
```

```python
dados['Valor Bruto'] = dados['Valor'] + dados['Condominio'] + dados['IPTU']
```

```python
dados
```

```python
dados['Valor do m2'] = dados ['Valor'] / dados ['Area']
```

```python
dados
```

```python
dados ['Valor do m2'] = dados ['Valor do m2'].round(2)
```

```python
dados
```

```python
dados['Valor Bruto do m2'] = (dados['Valor Bruto'] / dados['Area']).round(2)
```

```python
dados
```

```python
casa = ['Casa', 'Casa de Condomínio', 'Casa de Vila']
```

```python
dados['Tipo Agregado'] = dados['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Apartamento')
```
